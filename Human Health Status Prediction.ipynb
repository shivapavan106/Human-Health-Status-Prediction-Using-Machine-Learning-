```python
# dectect the outliers using boxplots 
for col in data.columns:
    if(data[col].dtype !='object'):
        sns.boxplot(data[col])
        plt.title(col)
        plt.show()
```

```python
# filling the missing values with mean(only numeric values)
numereic_columns = data.select_dtypes(include='number').columns
for i in numereic_columns:
    data[i] = data[i].fillna(data[i].mean())
```

```python
data.isnull().sum()
```

```python
# filling the missing values with mode( only categorical values)
cate_columns = data.select_dtypes(include='object').columns
for i in cate_columns:
    data[i] = data[i].fillna(data[i].mode()[0])
```

```python
# encoding the categories columns 
# they are types of encoding 
# 1. label encoding
# 2. one hot encoding
label_encode = data.select_dtypes(include='object').columns
for i in label_encode:
    print(f"{i} : {data[i].unique()}")

```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in data.columns:
    if(data[i].dtype == 'str'):
        data[i] = le.fit_transform(data[i])
```
